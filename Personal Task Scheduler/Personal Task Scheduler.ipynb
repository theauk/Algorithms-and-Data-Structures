### Types of Multitasking Tasks and Pairing
There are two types of multitasking tasks:
 
1. Multitasking tasks with dependencies (MT D).
2. Multitasking tasks without dependencies (MT%D).
 
A multitasking task could be taking the bus since it is dependent on walking to the bus stop. On the other hand, a multitasking task without dependencies could be reading about the palace on my phone because I can do that at any time. 
 
I need to determine an appropriate way to pair multitasking tasks. In this case, there are three possibilities as the order does not matter:
 
1. (MT D) and (MT D)
2. (MT%D) and (MT%D)
3. (MT D) and (MT%D)
 
We cannot pair taking the bus with taking another bus, which means that “(MT D) and (MT D)” should not be allowed to happen. Similarly, I cannot both read about the palace and the market at the same time resulting in “(MT%D) and (MT%D)” not being allowed either. Therefore, we can only combine a multitasking task with dependencies with a multitasking task without dependencies “(MT D) and “(MT%D).” 
 
### Partitioning
We can partition our time in different ways when executing multitasking tasks. We could finish the first multitasking task and then complete the next multitasking task. However, this would defeat the purpose of multitasking as we would be doing the tasks on their own. We could also decide to switch between two tasks. For example, I could ride the bus for a bit, read about the palace for a bit, ride the bus for a bit, etc. Similarly, the latter method also defeats the purpose of multitasking. 
 
Partitioning tasks that can be done at the same time would also worsen the time complexity. If we have a total time of two tasks, $n$, and a partitioning happening every, $t$, minutes. Then we would get that our running time would increase by:
 
$(n // t) * O(1)$
 
We can neglect the t as we get a large value of n which would then give us an additional running time of:
 
$O(n)$
 
Therefore, it makes sense to multitask if we want to optimize our time. 

### Ways to Decrease Overall Time

We can still consider different ways to improve our multitasking performance. With two multitasking tasks, there can be three cases:
 
1. They have the same time. 
2. MT D time > MT%D time.
3. MT D time < MT%D time.
 
 
When they have the same time, we would go ahead and start and end both tasks at the same time. 
 
For case 2, where MT D time > MT%D time, we could do the two at the same time and then finish MT D after MT%D is done. However, when we finish the MT%D task, we could bring in another MT%D task to optimize our time. After doing so, we would update the new MT%D’s time and status so that we can finish it later. For example, if I need to take the bus for 45 min (MTD), read about the palace for 30 min (MT%D) and read about the market for 30 min (MT%D) then I would:
 
1. Take the bus while reading about the palace for 30 min.
2. Continue taking the bus for 15 min while now reading about the market for 15 min. 
 
After this, the bus ride and reading about the palace would be marked as completed. Reading about the market would be marked as in-progress, and its time would be updated to 15 minutes (30 min – 15 min). 
 
For case 3, where MT D time < MT%D time, we would start both the MT D task and the MT%D task. However, we cannot finish the MT%D task, so we need to update its status and time to reflect the remaining time left. For example, if taking the bus takes 10 min and reading takes 15 min, then I would:
 
1. Take the bus for 10 min while reading for 10 min. 
 
When this is done, I will mark the bus ride as executed. The reading status would be set to in progress while the reading time would now be 5 min (15 min – 10 min). 
 
Overall, my algorithm will only match a non-dependency multitasking task with a dependency multitasking task because it helps me avoid clashes between tasks that cannot be done at the same time. Furthermore, if possible, I will do several multitasking tasks inside another task because this will decrease my overall time. 

```python
# From 3.2
def left(i):
    return(2*i + 1)

# From 3.2
def right(i):
    return(2*i + 2)
```

```python
# From 3.2
def max_heapify(heap, i):
    largest = i # Initializing the variable which stores the index with the largest element.
    
    # Finding the left and right indices using the functions above. 
    l = left(i)
    r = right(i)
    
    # If we still have have not analyzed the full heap and the element at l is greater than
    # the current largest element, we alter the index for the largest element. Otherwise, we
    # assign the index i as the index for the largest element.
    # We also have to use < and not <= because of Python indices starting at 0. Otherwise, we would
    # face a "list index out of range" in question 3.  
    if l < len(heap) and heap[l] > heap[i]: 
        largest = l
    else:
        largest = i
    
    # Doing the same for the right side of the binary tree. 
    if r < len(heap) and heap[r] > heap[largest]:
        largest = r
    
    # We have to check if we need to swab elements which we do when the largest element index is not equal to the
    # current index (i). If they are equal, then it means that the current element already is the largest one 
    # and thus we should not swab. 
    if largest != i:
        
        # Exchange the element at heap[i] with the element at heap[largest].
        heap[i], heap[largest] = heap[largest], heap[i] 
        
        # Calling heapify recursively on further subtrees. 
        max_heapify(heap, largest)
```

```python
def min_heapify(heap, i):
    
    smallest = i # Initializing the variable which stores the index with the smallest element.
    
    # Finding the left and right indices using the functions from question 2b. 
    l = left(i)
    r = right(i)
    
    # If we still have have not analyzed the full heap and the element at l is less than
    # the current largest element, we alter the smallest element's index. Otherwise, we
    # assign the index i as the index for the smallest element.
    if l < len(heap) and heap[l] < heap[i]: 
        smallest = l
    else:
        smallest = i
    
    # Doing the same for the right side of the binary tree. 
    if r < len(heap) and heap[r] < heap[smallest]:
        smallest = r
    
    # We have to check if we need to swab elements which we do when the smallest element index is not equal to the
    # current index (i). If they are equal, then it means that the current element already is the smallest one 
    # and thus we should not swab. 
    if smallest != i:
        
        # Exchange the element at heap[i] with the element at heap[largest].
        heap[i], heap[smallest] = heap[smallest], heap[i] 
        
        # Calling heapify recursively on further subtrees. 
        min_heapify(heap, smallest)
```

```python
# Heappush function for pushing an element into a heap at the root.
def heappush(heap, task):
    heap.insert(0, task) 
```

```python
# Popping an element from the root of a heap. 
# The function also saves the popped element so that we can update/reuse it.
def heappop(heap):
    current = heap[0]
    del heap[0]
    return current
```

```python
# Min heapifying the whole heap. 
def min_heapify_all(heap):
    for i in range((len(heap)-1) // 2, -1, -1):
        min_heapify(heap, i)
```

```python
# Max heapifying the whole heap. 
def max_heapify_all(heap):
    for i in range((len(heap)-1) // 2, -1, -1):
                max_heapify(heap, i)
```

```python
# Finding the task in a list/heap with the maximum number of dependencies.
def max_dependencies(tasks, current_max):
    
    # Going through all of the tasks in the list.
    for task in tasks:
        
        # If the length of the current task's dependency list is equal or longer than 
        # the current max, then we update the current max.
        if len(task[4]) >= current_max: 
            current_max = len(task[4])
            current_max_task = task
    
    return current_max_task
```

```python
# Function which deletes a task from the original list when we have pushed the task into the heap.
def delete_pushed_task(original_list, base_task):
    
    # Going through the tasks in the original task list.
    for task in original_list:
        
        # If the task is equal to the task we want to delete
        # we find its index and then delete it. 
        if task == base_task:
            delete_id = original_list.index(base_task)
            del original_list[delete_id]
```

```python
# Function which scores the tasks in the multitasking list.
def multitasking_scorer_heap(tasks, heap, no_depen_multi):
    
    # Initializing the score counter and sublist number (as the multitasking list
    # has nested lists corresponding to clusters of tasks).
    mt_score = 0
    sublist_number = 0
    
    # For every sublist in the list.
    for sublist in no_depen_multi:
        
        # For every task in each sublist we check count the number of tasks that
        # are dependent on it in the main heap.
        for mt_task in sublist:
            for element in heap:
                if mt_task[1] in tasks[element[1]][4]:
                    mt_score += 1

            # If the multitasking score has been updated, we update the current multitasking
            # tuple so that it has the new score.
            if mt_score > 0:
                new_tuple = (mt_score, mt_task[1])
                sublist.remove(mt_task)
                heappush(sublist, new_tuple)
                mt_score = 0
            
            # When we have checked the whole sublist we move to the next one. 
            sublist_number += 1
```

```python
# Function which pushes dependencies of a task either into the heap or the no_dep_multi tasking list.
def push_dependency(heap, tasks, original_list, task, no_depen_multi):
    
    # Grabbing the global values we need from the main function. 
    global length
    global priority_score
    global no_depen_multi_counter
    
    # Appending a new empty list for the current cluster of tasks (dependencies of a max
    # dependency task).
    no_depen_multi.append([])
    
    # Going through the dependencies in the specific input task.
    for dependencies in task[4]:
        
        # If task dependencies > 0, we push it into the main heap and delete it from the original list. 
        if len(tasks[dependencies][4]) != 0:
            heappush(heap, (priority_score + len(tasks[dependencies][4]), dependencies))
            delete_pushed_task(original_list, tasks[dependencies])
            min_heapify_all(heap)
        
        # If task dependencies = 0 and multitasking = false, we push it into the main heap with a 
        # score of 0 and then delete it from the original list.
        elif tasks[dependencies] in original_list and tasks[dependencies][3] == False:
            heappush(heap, (0, dependencies))
            min_heapify_all(heap)
            delete_pushed_task(original_list, tasks[dependencies])
            
        # If task dependencies = 0 and multitasking = true, we push it into the no_dep_multi tasking list.
        elif tasks[dependencies] in original_list and tasks[dependencies][3] == True:
            heappush(no_depen_multi[no_depen_multi_counter], (0, dependencies))
            delete_pushed_task(original_list, tasks[dependencies])
            
            for sublist in no_depen_multi:
                max_heapify_all(sublist)
    
    # Updating the score and counter. 
    priority_score -= length
    no_depen_multi_counter += 1
```

```python
# Function which updates our schedule for non-multitasking tasks. 
def update_schedule(schedule, tasks, task, start_time, counter):
    
    # Finding the start hour and min (as start time is given in totak min).
    # E.g. 9:00 = 9*60 = 540 min.
    start_hours = start_time // 60
    start_min = start_time % 60
    
    # Finding the end hour and min. 
    end_hours = (start_time + task[2]) // 60
    end_min = (start_time + task[2]) % 60
    
    # Increasing the start time by the task duration.
    start_time += task[2]
    
    # Appending the result to the schedule in the right format.
    task_time = "Start: {}:{}, End: {}:{}".format(start_hours, start_min, end_hours, end_min)
    schedule.append([counter, task[1], task_time])

    # Returning the updated start time. 
    return start_time
```

```python
# Function which updates the schedule for multi-tasking tasks.
# We need a different one because when a multitasking is performed we use the time from an element in 
# one list for a different element. However, we still want the original name of the latter task. 
def update_schedule_multi(schedule, tasks, task_time, task_name, start_time, counter):
    
    # Finding start time.
    start_hours = start_time // 60
    start_min = start_time % 60
    
    # Finding end time.
    end_hours = (start_time + task_time[2]) // 60
    end_min = (start_time + task_time[2]) % 60
    
    # Updating start time.
    start_time += task_time[2]
    
    # Updating the schedule .
    task_time = "Start: {}:{}, End: {}:{}".format(start_hours, start_min, end_hours, end_min)
    # Here we use different variables than the function above to correctly create the list entry.
    schedule.append([counter, task_name[1], task_time]) 

    # Returning the updated start time. 
    return start_time
```

```python
# Function which deletes/updates the dependencies in the main task list after a task has been executed. 
def delete_dependencies(tasks, base_task_id):
    
    # Go through every task.
    for task in range(len(tasks)):
        
        # Examine a task's dependencies
        for dependency in tasks[task][4]:
            
            # If the base task's id is in the dependency list we
            # delete it because the base task has been executed.
            if base_task_id == dependency:
                tasks[task][4].remove(base_task_id)
```

```python
# Function which ensures that we are executing tasks properly and put possible multitasking tasks together. 
def multitasking(heap, tasks, no_depen_multi, popped_task, schedule, start_time, counter):
    
    # If the popped task is not a multitasking task.
    if popped_task[3] == False:
        
        # If all the dependencies have been executed.
        if len(popped_task[4]) == 0:
            
            # Schedule the task as it is ready to be executed but only do it if time > 0.
            if popped_task[2] != 0:
                start_time = update_schedule(schedule, tasks, popped_task, start_time, counter)
                
                # Update the counter
                counter += 1
            
            # Updating the time and the status of the task.
            tasks[popped_task[0]][2] = 0
            tasks[popped_task[0]][5] = "COMPLETED"
            
            # Delete it from the dependency list of other tasks.
            delete_dependencies(tasks, popped_task[0]) 
        
        # If there are missing dependencies for the popped task:
        # Due to the nature of the scheduler we know that the missing dependencie(s) is/are in the
        # multitasking no dependency list. Thus, if there are more than one missing dependency then we also 
        # know that it doesn't matter which order they are executed since all of them have no dependencies.
        # Furthermore, we cannot schedule two missing dependencies at the same time because their characteristics
        # do not allow that.
        else:
            
            # Going through all dependencies in the popped task.
            for dependency in popped_task[4]:
                
                # Finding the current root value in the heap.
                current_min_value = heap[0][0]
                
                # Pushing the dependencies into the heap with a smaller value than the root. 
                # So that we know they will get on top of the heap.
                heappush(heap, (current_min_value - 1, dependency))
                
                # Heapify
                min_heapify_all(heap)
                
                #Delete it::::: Because if it could have been paired with something else it would have been earlier. 
                #But now we're at a point where we have to do it no matter what so we can put it in the heap.
                #And even if we push more than one dependency from the MT no dep list then they cannot be done 
                #together anyways.
                
                # We delete the element from the multitasking sublist as it is now in the heap.
                # We can do so because we know that the multitasking task does not have any dependencies left
                # because then it would not have been in the multitasking list. Thus, we know that it will
                # get executed as the next task and we do not need to refer to it in the multitasking task list.
                for sublist in range(len(no_depen_multi)):
                    for the_task in no_depen_multi[sublist]:
                        if dependency == the_task[1]:
                            no_depen_multi[sublist].remove(the_task)
                            
  
    # If the popped task is a multitasking task. 
    else:
        
        # If all the dependencies have been executed, we can combine it with a task from the no-dependency
        # multitasking list.
        if len(popped_task[4]) == 0:
            
            # If there are no multitasking-no-dependency tasks left. Then we go ahead
            # and execute the current popped task.
            if len(no_depen_multi) == 0:
                
                # Schedule the task as it is ready to be executed unless its time is already 0.
                if popped_task[2] != 0:
                    start_time = update_schedule(schedule, tasks, popped_task, start_time, counter)
                    
                    # Update the counter
                    counter += 1

                # Updating the time and the status of the task.
                tasks[popped_task[0]][2] = 0
                tasks[popped_task[0]][5] = "COMPLETED"

                # Delete it from the dependency list of other tasks.
                delete_dependencies(tasks, popped_task[0])
                
            # If there are tasks left in the multitasking no-dependency list, we combine the popped
            # task with a task from this list.
            else:
                
                # Pop a task from the multitasking no-dependency list.
                popped_multi = heappop(no_depen_multi[0])
                popped_multi_task = tasks[popped_multi[1]]
                
                # Finding the current start time.
                start_time_same = start_time
                
                # If the multitasking task and the popped task take the same time.
                # We complete both tasks with the same start and end time. 
                if popped_task[2] == popped_multi_task[2]:
                
                    # Schedule the tasks as they are ready to be executed.
                    if popped_task[2] != 0 and popped_multi_task[2] != 0:
                        update_schedule(schedule, tasks, popped_task, start_time_same, counter)
                        start_time = update_schedule(schedule, tasks, popped_multi_task, start_time_same, counter)

                        # Update the counter
                        counter += 1

                    # Updating the time and the status of the tasks.
                    tasks[popped_task[0]][2] = 0
                    tasks[popped_task[0]][5] = "COMPLETED"

                    tasks[popped_multi_task[0]][2] = 0
                    tasks[popped_multi_task[0]][5] = "COMPLETED"

                    # Delete the tasks from the dependency list of other tasks.
                    delete_dependencies(tasks, popped_task[0])
                    delete_dependencies(tasks, popped_multi_task[0])
                
                # If the popped task takes more time than the multitasking task.
                # We complete part of the popped task and the full multitasking task. 
                # Then we push the remainder of the popped task into the heap again and change its status.
                elif popped_task[2] > popped_multi_task[2]:
                    
                    # Finding the time difference.
                    time_difference = popped_task[2] - popped_multi_task[2]
                    
                    # Schedule the multitasking task.
                    if popped_multi_task[2] != 0:
                        update_schedule(schedule, tasks, popped_multi_task, start_time_same, counter)
                        
                    # Scheduling part of the popped task.
                    # We schedule it with the time of the multitasking task.
                    if popped_task[2] != 0:
                        start_time = update_schedule_multi(schedule, tasks, popped_multi_task, popped_task, start_time, counter)
                        
                    # Update the counter
                    counter += 1
                    
                    # Update the tasks.
                    tasks[popped_multi_task[0]][2] = 0
                    tasks[popped_multi_task[0]][5] = "COMPLETED"
                    tasks[popped_task[0]][2] = time_difference
                    tasks[popped_task[0]][5] = "IN-PROGRESS"
                    
                    # Push the popped task back to the heap with the time difference unless the 
                    # left-over time is now zero. 
                    current_min_value = heap[0][0]
                    if tasks[popped_task[0]][2] != 0:
                        heappush(heap, (current_min_value - 1, popped_task[0]))
                    
                    # Delete the multitask task from the dependency list of other tasks.
                    delete_dependencies(tasks, popped_multi_task[0])
                
                # If the popped task takes less time than the multitasking task.
                # We complete part of the multitasking task and the full popped task. 
                # Then we push the remainder of the multitasking task into the multitasking list again 
                # and change its status.
                elif popped_task[2] < popped_multi_task[2]:
                    
                    # Finding the time difference. 
                    time_difference = popped_multi_task[2] - popped_task[2]
                    
                    # Schedule the popped task.
                    if popped_task[2] != 0:
                        update_schedule(schedule, tasks, popped_task, start_time_same, counter)
                    
                    # Scheduling part of the multitasking task and pushing the remainder back into the multitasking
                    # list. We schedule it with the time of the popped task.
                    if popped_multi_task[2] != 0:
                        start_time = update_schedule_multi(schedule, tasks, popped_task, popped_multi_task, start_time, counter)
                    
                    # Push the popped multitasking task back to the multitasking list with the time difference. 
                    current_min_value = heap[0][0]
                    heappush(no_depen_multi[0], (current_min_value - 1, popped_multi_task[0]))
                    
                    # Update the tasks.
                    tasks[popped_task[0]][2] = 0
                    tasks[popped_task[0]][5] = "COMPLETED"
                    tasks[popped_multi_task[0]][2] = time_difference
                    tasks[popped_multi_task[0]][5] = "IN-PROGRESS"
                    
                    
                    # Delete the popped task from the dependency list of other tasks.
                    delete_dependencies(tasks, popped_task[0])
                    
                    # Update the counter
                    counter += 1
                 
        # Here, we have a multitasking task which has dependencies left. Due to the nature of the heap, 
        # if this multitasking task was dependent on any other task in the heap, it would not be at 
        # the top of the heap. Therefore, this task is not dependent on any task that comes after it in the
        # heap. The latter means that it is dependent on a task in the non-dependent multitasking list. 
        else:
                 
            # Going through all dependencies in the popped task.
            for dependency in popped_task[4]:
                
                # Finding the current root value in the heap.
                current_min_value = heap[0][0]
                
                # Pushing the dependencies of the popped task into the heap with a smaller value than the root. 
                # So that we know they will get on top of the heap.
                heappush(heap, (current_min_value - 1, dependency))
                
                # Heapify
                min_heapify_all(heap)
                
                
                #We delete the element from the multitasking sublist as it is now in the heap.
                for sublist in range(len(no_depen_multi)):
                    for the_task in no_depen_multi[sublist]:
                        if dependency == the_task[1]:
                            no_depen_multi[sublist].remove(the_task)

    # Returning the start time and counter.
    return(start_time, counter)  
```

```python
# Function which returns the final schedule. 
def scheduler(tasks, heap, no_depen_multi):
    
    # Creating an emoty schedule and starting the counter which will
    # indicate the order of the tasks. 
    schedule = []
    counter = 1
    
    # Setting the start time at 9:00 am.
    start_time = 9*60
    
    # While there are still tasks left in either of the heaps.
    while len(heap) > 0 or len(no_depen_multi) > 0:
        
        # Deleting empty sublists in the multitask list to avoid list index errors.
        no_depen_multi = [x for x in no_depen_multi if x != []]
        
        # Pop the top element in the main heap.
        popped_task = heappop(heap)
        
        # Fetch the full task info of the popped element.
        task = tasks[popped_task[1]]
        
        # Plan tasks while also taking multitasking and dependencies into account.
        result_multi = multitasking(heap, tasks, no_depen_multi, task, schedule, start_time, counter)
        
        # Updating the start time and the counter.
        start_time = result_multi[0]
        counter = result_multi[1]
        
        # Min heapify the main heap. 
        min_heapify_all(heap)
    
    # Returning the final schedule. 
    return schedule 
```

```python
# The main function. 
def task_scheduler(tasks):
    
    # Create overall heap.
    heap = [] 
    
    # Create a global variable for length and priority score. 
    global length
    length = len(tasks)
    
    global priority_score
    # Here we set the value to an appropriate amount dependening on the number of tasks.
    # As we are only examining tasks for a single day, 1000 is definitely sufficient.
    priority_score = 1000 
    
    # Counting which no-dependency multitasking list we should currently add elements too.
    global no_depen_multi_counter
    no_depen_multi_counter = 0
    
    # Creating a list for no-dependency multitasking tasks. 
    no_depen_multi = []

    # Create a copy of the task list. 
    original_list = []
    
    # We need to make a "deep copy" of the task list so that changes to the copied 
    # list does not affect the task list and vice versa. 
    for i in tasks:
        original_list.append(i)
    
    # While we have not placed all elements in the heap we continue.
    while len(original_list) > 0:
        
        # Resetting the max dependencies variable. We use -1 as we know that all tasks will
        # have at least 0 dependencies. 
        current_max_dependencies = -1
        
        # Finding the task with the max dependencies
        current_max_dependencies = max_dependencies(original_list, current_max_dependencies) 
        
        # If it is a non-dependency multitasking task.
        if len(current_max_dependencies[4]) == 0 and current_max_dependencies[3] == True:
            
            # If the length of the non-dependency multitasking list is not equal to zero
            # we push the element into the first index of the first sublist.
            if len(no_depen_multi) != 0:    
                heappush(no_depen_multi[0], (0, current_max_dependencies[0]))
                delete_pushed_task(original_list, current_max_dependencies)
            
            # Otherwise, we push it into the list.
            else:
                heappush(no_depen_multi, (0, current_max_dependencies[0]))
                delete_pushed_task(original_list, current_max_dependencies)
            
            # Max-heapifying the dependency sublists (as a greater score means higher priority).
            for sublist in no_depen_multi:
                max_heapify_all(sublist)
        
        # For the rest of the tasks.
        else:
            # Pushing the current task into the heap.
            heappush(heap, (priority_score + len(current_max_dependencies[4]), current_max_dependencies[0]))

            # Finding the max index
            max_index = original_list.index(current_max_dependencies)

            # Deleting it from the list with tasks not prioritized yet.
            delete_pushed_task(original_list, original_list[max_index])

            # Push dependencies of max task into the heap
            push_dependency(heap, tasks, original_list, current_max_dependencies, no_depen_multi)
    
    # Min-heapifying the main task heap.
    min_heapify_all(heap)
    
    # Prioritize the multitasking no-dependency tasks.
    multitasking_scorer_heap(tasks, heap, no_depen_multi)
    
    # Produce a schedule
    final_schedule = scheduler(tasks, heap, no_depen_multi)
    
    # Print the final schedule with a line break between each task.
    for task in final_schedule:
        print(task)
```

## Explanation of Algorithm
The algorithm first clusters tasks together in a min heap based on activities with the ones with more dependencies further down the heap. Tasks with zero dependencies are placed at the beginning, and no-dependency multitasking tasks are placed in a separate max heap. When an activity has been placed in a heap, we delete it from a copy of our task list so that we can keep track of when to start further prioritization. 

Next, it prioritizes the tasks in the no-dependency multitasking list based on the length of the dependency lists. Furthermore, tasks belonging to an activity are inside the same nested list, and these nested lists are ordered in the same way as the main task heap. As a result, this will optimize our time as we pop out multitasking tasks related to the current activity first. 

Afterward, we start to pop tasks out of the heap and put them into the schedule list. We do this by having different if-elif-else statements that check if it is a multitasking task and if it has dependencies left. 

If it is ready to be executed and cannot be multitasked, then we perform it. If it cannot be multitasked but has dependencies left, we place these dependencies in the heap before the main task so that they will be executed next. 

If it can be multitasked, we check if it has dependencies we need to execute first. If it has, we push those into the heap and execute them. Otherwise, we check if it can be paired with a task from the non-dependency multitasking list. If yes, then we start pairing activities in the way described in part C, which would optimize the time taken. 

Furthermore, we both min-heapify the main task heap and max-heapify the non-dependency multitasking list several places in the algorithm to ensure that the next element we pop has the greatest priority. Throughout the process, the algorithm also ensures to update the times, dependency lists, and statuses while also creating the output schedule list. 

## Explanation of Output
The algorithm outputs a schedule with the start and the end time of each task. The number at the beginning of the task indicates the order of each task. If we have two of the same number immediately after each other, then it means that we should multitask those tasks. The latter is also reflected in the start time of multitasking tasks as it is the same for both. Furthermore, a task might be visible more than once. This happens if we partition the task between different multitasking tasks or when we have to finish a task which we did not complete earlier (a task which is still "in-progress"). 
 
## Benefits
 
### Outputs Proper Schedule
After understanding the format of the output and checking the schedule, we can see that the scheduler outputs a list that would make sense to follow. For the output schedule, activity tasks are clustered together, we multitask only between MTD and MT%T tasks, times are updated, and the executing number to the left correctly identifies multitasking (when a number is there twice) or no multitasking (when a number is there once). Furthermore, we stay inside a single day with a starting time at 9 am and ending time at 9:48 pm, and these two times are also realistic in terms of a regular day schedule.
 
### Clustering tasks and Ascending Order
Initially, I had difficulties with clustering tasks for different activities together and considered adding an input reflecting a task's activity or chaining all tasks. For example, scoring tasks based on dependencies would ask me to take a bus to the palace and then another bus to the market, although I should visit the palace before leaving for the market. The latter happened because they had the same amount of dependencies. 
 
Instead, the algorithm now clusters tasks depending on which activity they belong to, which ensures that the schedule makes sense. This is possible because the algorithm picks a task with the highest number of dependencies and then places it in the main heap. Next, it scores the task's dependencies based on its number of dependencies and puts it in the heap. After this, we lower a counter and repeat the process. Thus, the next cluster of tasks would get scores lower than the tasks belonging to a different activity. 
 
Therefore, a person can also use the scheduler if they want to perform activities based on an ascending order of max task dependencies. 
 
### Multitasking
Another strength of the algorithm is that it takes advantage of our ability to multitask, and it does so without a clash between multitasking tasks that cannot be done together. In the algorithm, we then only multitask when we have a multitasking task with no dependencies and a multitasking task with dependencies. For example, I might as well prepare the hiking route while I am commuting to Isadong. Thus, implementing the algorithm to allow for this type of multitasking saves time.
 
Furthermore, the algorithm does not just pick the first available multitasking task. Instead, it first clusters tasks together in a nested list depending on the activity they belong to, and this order corresponds to the order in the main task heap. Second, inside the sub-lists, the tasks have been ranked based on how many dependencies there are on them. Thus, when we pop the first non-dependency multitasking task, we know that it will be a task that should have higher priority given the format of the main heap. 
 
## Drawbacks
 
### Criteria for Prioritizing Activities
In my algorithm, the last task we execute will be the one that has the task with the max length of its dependency list. Therefore, the last activity we complete is the activity, which includes the latter task. However, this might not be the optimal way to schedule our tasks. 
 
For example, I might have two activities: going to the palace and going to the market. In this case, going to the market might have the task with the longest dependency list, which would push the activity to the end of the list, and then we would perform the activities related to the palace first. Still, if the palace has fewer but way longer dependencies timewise than the market's dependencies, then we might be better off going to the market and then the palace. The latter is the case because when we execute the tasks related to the market we might have more opportunities to multitask and thus finish the dependencies for the palace. Thus, switching the activities might improve our overall time. 
 
Overall, the algorithm's priority assignment means that there might be potential to optimize the time if we have an input similar to the example with the palace and the market. However, to account for this, we would need a more sophisticated algorithm that performs more calculations and comparisons.
